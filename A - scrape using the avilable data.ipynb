```python
import pandas as pd
import numpy as np
import requests
from bs4 import BeautifulSoup
import json
import time
from time import strftime
from datetime import datetime
```

# Get Meta data

We have downloaded the metadata for 200000 campaings from https://webrobots.io/kickstarter-datasets/, and this notebooks puts all that data into a dataframe

```python
import glob
path ='/Kickstarter_2018-02-15T03_20_44_743Z' # use your path
allFiles = glob.glob(path + "/*.csv")
frame = pd.DataFrame()
list_ = []
i=1
for file_ in allFiles:
    print(i, end=" ")
    df = pd.read_csv(file_,index_col=None, header=0,usecols=columns)
    print(len(df),end="--")
    list_.append(df)
    i+=1
frame = pd.concat(list_,ignore_index=True)
```

```python
len(frame)
```

# Now I just want the ones in the US

```python
json.loads(frame["location"][0])
```

```python
def my_json_loads(x,field):
    try:
        return json.loads(x)[field]
    except:
        return {}
```

```python
countries=frame["location"].apply(my_json_loads,args=("country",))
states=frame["location"].apply(my_json_loads,args=("state",))
#frame["category"]=frame["category"].apply(my_json_loads)
```

```python
frame["country"]=countries
frame["state"]=states
```

```python
frame=frame.drop(columns=["location"]);
```

```python
frame=frame[frame["country"]=="US"]
```

# Create a category - subcategory field

```python
json.loads(frame["category"][0])
```

```python
categories=frame["category"].apply(my_json_loads,args=("slug",))
```

```python
category=categories.str.split('/').str[0]
subcategory=categories.str.split('/').str[1]
```

```python
frame["category"]=category
frame["subcategory"]=subcategory
```

# convert dates

```python
frame["created_at"]=frame["created_at"].apply(datetime.fromtimestamp)
frame["deadline"]=frame["deadline"].apply(datetime.fromtimestamp)
```

# Urls convertions

```python
def nested_dic(x,field):
    try:
        return x[field]
    except:
        {}
```

```python
json.loads(frame["urls"][0])["web"]['project']
```

```python
links=frame["urls"].apply(my_json_loads,args=("web",)).apply(nested_dic,args=("project",))
```

```python
frame["link"]=links
```

```python
frame=frame.drop(columns="urls")
```
